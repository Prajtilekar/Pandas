drop_duplicate()

```python
s.drop_duplicates()
```

```python
p = pd.Series(np.random.randint(1,10,10))
p
```

```python
p.drop_duplicates()
```

In drop_duplicates, it keeps 1st occurence and 2nd occurence will be deleted.\
Also inplace is False bydefault so changes are temporary, keep inplace is True so changes will become permnant.

```python
p
```

```python
p.drop_duplicates(keep='last')
```

```python
p #bydefault 1st occurace kept but we can change it by keep last, so last element remains and previous will be deleted.
```

```python
#If we want to remove all duplicates.
p
```

```python
p.drop_duplicates(keep=False)  
#All duplicates are removed.
```

#Series : - map()

```python
nm = pd.Series(['Male','Female','Male','Female','Female'])
nm
```

```python
nm.map({'Male':0}) #Here, with map we can't replace only one values if we do so other value become Nan 
#to avoid that if we want only one values changed use replace.
```

```python
# PS: convert Male to 0 and female 1
# Map values of Series according to input correspondence.
nm.map({'Male':0,'Female':1})
```

## Series:- replace()

```python
# replace Male as 0 and Female as 1
nm.replace('Male',0)
```

```python
nm.replace({'Male':0,'Female':1})
```
