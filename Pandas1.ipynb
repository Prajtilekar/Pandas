```python
#Syntax : s.apply(supply function)
a
```

```python
d=pd.Series(['aa','vf','def'])
d
```

```python
# Convert name into uppercase
d.apply(lambda nm:nm.upper())
```

```python
# Convert name into its length
d.apply(lambda name:len(name))
```

Use strip method to remove whitespaces

```python
v = pd.Series(['Narayan Rane','Ajit Pawar'])
v.apply(lambda n:n.split()[0])
```

```python
pin = pd.Series(['Ap.Satara Patan 542451','Pune Katraj 234567'])
pin
```

```python
pin.apply(lambda n:n[-6:])
```

```python
# using normal function
def fetchpin(a):
  return a[-6:]
pin.apply(fetchpin)
```
