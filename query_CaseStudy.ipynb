```python
#Query using df

df = pd.DataFrame({
                    'A':range(1,6),
                    'B':range(10,0,-2),
                    'C':range(10,15)   
})
df
```

```python
#Query the columns of a DataFrame with a boolean expression
df.query('A>B')
# this will return a df where element from column A > element from column B
```

```python
df.query('A<B')
#It will return rows where A is less than B
```

```python
df.query('A==3')
```

```python
#Without query how to fetch
df[df['A']==3]
```

```python
#fetch the value from column B where value of column A==3
df[df.A==3].B
```

```python
df[df.A==3]['B']
```

```python
df[df.A==3][['B']]
```

```python
df.query('A==3').B
```

```python
#fetch B and C where A==3
df.query('A==3')[['B','C']]
```
