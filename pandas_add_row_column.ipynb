## **Manipulation of DataFrame: How to add rows**

```python
import pandas as pd
```

```python
# List of Tuples
students = [ ('Jack', 34, 'Sydeny' , 'Australia') ,
             ('Riti', 30, 'Delhi' , 'India' ) ,
             ('Vikas', 31, 'Mumbai' , 'India' ) ,
             ('Neelu', 32, 'Bangalore' , 'India' ) ,
             ('John', 16, 'New York' , 'US') ,
             ('Mike', 17, 'las vegas' , 'US')  ]
students             
```

**Create DataFrame using tuples**

```python
df = pd.DataFrame(students, columns = ['Name','Age','City','Country'])
#df = pd.DataFrame(students, columns = ['Name' , 'Age', 'City' , 'Country'], index=['a', 'b', 'c' , 'd' , 'e' , 'f']),if index given manually
df
```

#**using dataframe.append() and Dictionary**

```python
# Pass the row elements as key value pairs thro append() function 
Out = df.append({'Name': "Prajakta",'Age':28}, ignore_index=True)
Out
```

**Here, ignore index is necessary because we dont have all columns values so it will ignore remaining and show it as NaN**

**here in Dictionary key act as column so no need to give columns separately**

```python
Dict1={'Name': "Praj",'Age':28}
Dict1
```

```python
df = df.append(Dict1,ignore_index=True)
df
```

```python
modDfObj = df.append({'Name' : 'Sunil' , 'Age' : 22,'pincode':123456} , ignore_index=True)
modDfObj
```

### **dataframe.append() and Series**

```python
#Create Series
s = pd.Series(['Raju', 21, 'Bangalore', 'India'], index = ['Name','Age','City','Country'])
s
```

```python
modDfObj = df.append(s, ignore_index=True)
modDfObj
```

```python
r = pd.Series(['Rohit',23,'Paris','US'])
r
# a Series without maping with columns
```

```python
df.append(r,ignore_index=True)
```

```python
r = pd.Series(['Rohit',23,'Paris','US'],index=['Name','Age','City','Country'])
r
# index mapping is an important 
```

```python
df.append(r,ignore_index=True)
```

```python
# list out column names/show column names
df.columns
```

```python
pd.Series(['Praj',22,'Pune','India'], index=df.columns)
```

```python
# List of series  
listOfSeries = [pd.Series(['Raju', 21, 'Bangalore', 'India'], index=df.columns ) ,
                pd.Series(['Sam', 22, 'Tokyo', 'Japan'], index=df.columns ) ,
                pd.Series(['Rocky', 23, 'Las Vegas', 'US'], index=df.columns ) ]
listOfSeries
```

```python
# Pass a list of series to the append() to add multiple rows
modDfObj = df.append(listOfSeries)
modDfObj
```

```python
#Deal with Index problems as follow:
modDfObj = df.append(listOfSeries, ignore_index=True)
modDfObj
```
